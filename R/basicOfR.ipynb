### Conditional statement in R

- A small program to check whether the number entered by user is even or odd.
  

```r
as.integer(readline(prompt = "Enter an Number :")) -> num
if(num %% 2 == 0){
    cat(num," is an even number")
}else{
    cat(num," is an odd number")
    }
```

### User defined function

```r
udfunc <- function(a)
{
    a*5 -> result
    print(result)
}
```

```r
udfunc(5)
```

```r
udfunc(69)
```

```r
udfunc2 <- function(x,y)
    {
    x+y -> result
    print(result)
    }
```

```r
udfunc2(5,5)
```

```r
udfunc2(60,9)
```

### Default Argument

```r
defArg <- function(i=10)
    {
    i*10 -> result
    print(result)
    }
```

```r
defArg()
```

```r
defArg(5)
```
