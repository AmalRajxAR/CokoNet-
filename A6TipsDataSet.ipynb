```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea 
import warnings 
warnings.filterwarnings('ignore')
```

### Loading the tips data set

```python
tipsdf = sea.load_dataset('tips')
tipsdf
```

## Working with tips data set using seaborn nd matplot

### 1) What are the count of the customers by genders ?

```python
sea.countplot( data = tipsdf, x = 'sex')
plt.show()
```

### 2) What are the count of the customers who smokes among different genders ?

```python
sea.countplot( data = tipsdf, x = 'sex', hue = 'smoker')
plt.show()
```

### 3) At what time do the customers usually vists ?

```python
sea.countplot( data = tipsdf, x = 'time')
plt.show()
```

### 4) At what time do customers of different gender usually vists?

```python
sea.countplot( data = tipsdf, x = 'time', hue = 'sex')
plt.show()
```

### 5) What is the distribution of the total bill amonunts?

```python
tipsdf.total_bill.plot.hist();
```

```python
tipsdf.total_bill.plot.kde();
```

### 6) Is there a relationship between the total bill and tip ?

```python
tipsdf.plot.scatter( x = 'total_bill', y = 'tip')
```

### 7) What is the average tip received on different days of the week ?

```python
sea.barplot( data = tipsdf, x = 'day', y = 'tip');
```

### 8) What is the count of the customers for each day of week ?

```python
sea.countplot( data = tipsdf, x = 'day');
```

### 9) What were the tips received according to the gender ?

```python
sea.boxplot( data = tipsdf, x = 'sex', y = 'tip');
```

### 10) How does the tip amount vary based on whether the customers was a smoker or not ?

```python
sea.boxplot( data = tipsdf, x = 'smoker', y = 'tip')
```

### 11) The fluctuation of the total bill across different days. 

```python
tipsdf.plot.line( x = 'day', y = 'tip', figsize = (20,5))
```

### 12) The difference in the tip amount received in each time.

```python
sea.violinplot( data = tipsdf, x = 'time', y = 'tip');
```

### 13) What is the percentage of the customer that visits in each day ?

```python
plt.pie( data = tipsdf, x = tipsdf.day.value_counts(), labels = ['Sat','Sun','Thur','Fri'], autopct = '%0.2f%%')
plt.show()
```
