# Filtering 

### This data shows the vehicles with fuel efficiency less than 20 

```python
cardf[cardf['Fuel_efficiency']>20]
```

### This data shows the vehicles made after the model named "Civic"

```python
cardf[cardf['Model'] == 'Civic']
```

### This data shows the vehicles that are not a passanger type

# GroupBy

### Average horsepower for the particular vehicle type

```python
cardf.groupby(['Vehicle_type'])['Horsepower'].mean().reset_index()
```

### Total price of the vehicles from the particular manufacturer

```python
cardf.groupby(['Manufacturer'])['Price_in_thousands'].sum().reset_index()
```

# .pivot() - Creates new index and column from the existing data

### Average horsepower of the vechicle type from each manufacturer 

```python
cardf.pivot_table(values='Horsepower', index='Manufacturer', columns='Vehicle_type', aggfunc="mean")
```

### Total sales of the cars from each manufacturer according to their vehicle type

```python
cardf.pivot_table(values='Sales_in_thousands', index='Manufacturer', columns='Vehicle_type', aggfunc='sum')
```

# Cut method in pandas

```python
hp = [0,150,300,500]
cartype = ['EconomyCars','SUV','SportsCars']
cardf['Speedy'] = pd.cut(cardf.Horsepower, bins=hp, labels=cartype)
```

```python
cardf.sample(5)
```

```python
cardf.Speedy.value_counts()
```

```python
cardf.groupby(['Speedy'])['Fuel_efficiency'].mean().reset_index()
```

# n_largest and n_smallest

```python
cardf.nsmallest(3,'Sales_in_thousands')
```
